### **Screen Time Project**



```python
import pandas as pd
```

```python
# Read in you weekly expenses dataset (Make sure it is a csv file)
link = 'https://raw.githubusercontent.com/kennethayson/Skyline-Math-200/main/Screen%20Activity%20-%20Kenneth%20Ayson%20-%20Sheet1.csv'

screen_df = pd.read_csv(link)
```

```python
screen_df['Minutes'].describe()
```

```python
import seaborn as sns
```

```python
sns.histplot(x='Minutes', data=screen_df, bins=5)
```

```python
screen_df ['Minutes'].mean().round()
```

```python
sns.histplot(x='Minutes', data=screen_df, bins=50)
```

```python
screen_df['Day_of_the_week'].value_counts().plot(kind='pie', autopct=' %1.1f%%')
```

```python
screen_df['App'].value_counts().head(3)
```

```python
screen_df['Category'].value_counts().plot(kind = 'pie')
```

```python
screen_df['Category'].value_counts(normalize = True).plot(kind = 'bar', color = 'orange', width = 0.5)
```

```python
screen_df['App'].value_counts(normalize=True).plot(kind='bar')
```

```python
screen_df['App'].value_counts().plot(kind = 'pie')
```

```python
screen_df['Category'].value_counts().head(3)
```

```python
screen_df.groupby('Category')['Minutes'].agg(['mean'])
```

```python
sns.boxplot(x='Category', y='Minutes', data=screen_df)
```

```python
screen_df.groupby('Day_of_the_week')['Minutes'].agg(['mean'])
```

```python
sns.boxplot(x='Day_of_the_week', y='Minutes', data=screen_df)
```
