### **Screen Time Project**



```python
import pandas as pd
```

```python
# Read in you weekly expenses dataset (Make sure it is a csv file)
link = 'https://raw.githubusercontent.com/kennethayson/Skyline-Math-200/main/Screen%20Activity%20-%20Kenneth%20Ayson%20-%20Sheet1.csv'

screen_df = pd.read_csv(link)
```

```python
screen_df['Minutes'].mean().round(2)
```

```python
import seaborn as sns
```

```python
sns.histplot(x='Minutes',data=screen_df)
```

```python
sns.boxplot(x = 'Minutes' , y = 'App', data = screen_df,  palette="Blues")
```

```python
sns.boxplot(x = 'Day_of_the_week' , y = 'Minutes', data = screen_df,  palette="Blues")
```

```python
sns.boxplot(x = 'Minutes' , y = 'Category', data = screen_df,  palette="Blues")
```
